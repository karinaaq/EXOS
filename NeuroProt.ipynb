# K- Means


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

datos = pd.DataFrame({
    'ang': y,
    'acc': z,
})

np.random.seed(1)
k = 4
centroids = {
    i+1: [np.random.randint(0, 80), np.random.randint(0, 750)]
    for i in range(k)
}

fig = plt.figure(figsize=(5,5))
plt.scatter(datos['ang'], datos['acc'], color='k')
colmap = {1:'r', 2:'g', 3:'b', 4:'m'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color = colmap[i])

plt.show()
```

```python
def assigment(datos, controids):
    for i in centroids.keys():
        datos['distance_from_{}'.format(i)] = np.sqrt((datos['ang']-centroids[i][0])**2 + (datos['acc']-centroids[i][1])**2)
    
    centroid_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    datos['cercano'] = datos.loc[:,centroid_distance_cols].idxmin(axis=1)
    datos['cercano'] = datos['cercano'].map(lambda x: int(x.lstrip('distance_from_')))
    datos['color'] = datos['cercano'].map(lambda x: colmap[x])
    return datos

datos = assigment(datos, centroids)
print(datos.head())

fig = plt.figure(figsize = (5,5))
plt.scatter(datos['ang'], datos['acc'], color = datos['color'], alpha = 0.5, edgecolor = 'k')
for i in centroids.keys():
    plt.scatter(*centroids[i], color = colmap[i])

plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
import pandas as pd
```

```python
datos = pd.DataFrame({
    'ang': y,
    'acc': z,
})
fig = plt.figure(figsize=(20,10))
plt.scatter(datos['ang'], datos['acc'], color='k')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(datos)
y_kmeans = kmeans.predict(datos)
```

```python
plt.scatter(datos['ang'], datos['acc'], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_

plt.scatter(centers[:, 0], centers[:, 1], c='black', s=100, alpha=0.5);
```

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(datos, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(datos.shape[0])[:n_clusters]
    centers = [
        (datos['ang'][i[1]], datos['acc'][i[1]]), (datos['ang'][i[2]], datos['acc'][i[2]]),
        (datos['ang'][i[3]], datos['acc'][i[3]]),(datos['ang'][i[0]], datos['acc'][i[0]])

    ]
    
    while True:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(datos, centers)
        
        # 2b. Find new centers from means of points
        new_centers = np.array([datos[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels

centers, labels = find_clusters(datos, 4)
plt.scatter(datos['ang'], datos['acc'], c=labels,
            s=50, cmap='viridis');
```

```python
centers, labels = find_clusters(datos, 4, rseed=0)
plt.scatter(datos['ang'], datos['acc'], c=labels,
            s=50, cmap='viridis');

```

## In Depth: Gaussian Mixture Models

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

def plot_kmeans(kmeans, datos, n_clusters=4, rseed=0, ax=None):
    labels = kmeans.fit_predict(datos)

    # plot the input data
    ax = ax or plt.gca()
    ax.axis('equal')
    ax.scatter(datos['ang'], datos['acc'], c=labels, s=40, cmap='viridis', zorder=4)

    # plot the representation of the KMeans model
    centers = kmeans.cluster_centers_
    radii = [cdist(datos[labels == i], [center]).max()
             for i, center in enumerate(centers)]
    for c, r in zip(centers, radii):
        ax.add_patch(plt.Circle(c, r, fc='#CCCCCC', lw=3, alpha=0.5, zorder=1))
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
plot_kmeans(kmeans, datos)
```

```python
rng = np.random.RandomState(13)
D = []
D = np.dot(datos, rng.randn(2, 2))
D = np.array(D).T
D = D.tolist()

D_stretched = pd.DataFrame({
    'ang': D[:][0],
    'acc': D[:][1],
})

kmeans = KMeans(n_clusters=4, random_state=0)
plot_kmeans(kmeans, D_stretched)
```

# Generalizing E–M: Gaussian Mixture Models

```python
from sklearn import mixture

gmm = mixture.GaussianMixture(n_components=4).fit(datos)
labels = gmm.predict(datos)
plt.scatter(datos['ang'], datos['acc'], c=labels, s=40, cmap='viridis');
```

```python
probs = gmm.predict_proba(datos)
print(probs[:5].round(3))
```

```python
size = 50 * probs.max(1) ** 2  # square emphasizes differences
plt.scatter(datos['ang'], datos['acc'], c=labels, cmap='viridis', s=size);
```

```python
from matplotlib.patches import Ellipse

def draw_ellipse(position, covariance, ax=None, **kwargs):
    """Draw an ellipse with a given position and covariance"""
    ax = ax or plt.gca()
    
    # Convert covariance to principal axes
    if covariance.shape == (2, 2):
        U, s, Vt = np.linalg.svd(covariance)
        angle = np.degrees(np.arctan2(U[1, 0], U[0, 0]))
        width, height = 2 * np.sqrt(s)
    else:
        angle = 0
        width, height = 2 * np.sqrt(covariance)
    
    # Draw the Ellipse
    for nsig in range(1, 4):
        ax.add_patch(Ellipse(position, nsig * width, nsig * height,
                             angle, **kwargs))
        
def plot_gmm(gmm, X, label=True, ax=None):
    ax = ax or plt.gca()
    labels = gmm.fit(X).predict(X)
    if label:
        ax.scatter(X['ang'], X['acc'], c=labels, s=40, cmap='viridis', zorder=2)
    else:
        ax.scatter(X['ang'], X['acc'], s=40, zorder=2)
    ax.axis('equal')
    
    w_factor = 0.2 / gmm.weights_.max()
    for pos, covar, w in zip(gmm.means_, gmm.covariances_, gmm.weights_):
        draw_ellipse(pos, covar, alpha=w * w_factor)
```

```python
gmm = mixture.GaussianMixture(n_components=4, random_state=42)
plot_gmm(gmm, datos)
```
