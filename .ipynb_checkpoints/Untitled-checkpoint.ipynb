```python
import pandas as pd
import numpy as np
import random
from time import time
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.patches import Arc
import plotly.express as px 
import csv
```

## Carga del Dataset 

```python
#------------------------------N1-------------------------------------
x=[]
y=[]
z=[]
with open('BaseDatos/N1.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
        z.append(int(row[2]))
d1 = pd.DataFrame({
    't': x,
    'ang': y,
    'acc': z,
})
#------------------------------N2-------------------------------------
x2=[]
y2=[]
z2=[]
with open('BaseDatos/N2.csv', 'r') as csvfile:
    plots2= csv.reader(csvfile, delimiter=',')
    for row2 in plots2:
        x2.append(int(row2[0]))
        y2.append(int(row2[1]))
        z2.append(int(row2[2]))
d2 = pd.DataFrame({
    't':x2,
    'ang': y2,
    'acc': z2,
})
#------------------------------N3-------------------------------------
x3=[]
y3=[]
z3=[]
with open('BaseDatos/N3.csv', 'r') as csvfile:
    plots3= csv.reader(csvfile, delimiter=',')
    for row3 in plots3:
        x3.append(int(row3[0]))
        y3.append(int(row3[1]))
        z3.append(int(row3[2]))
d3 = pd.DataFrame({
    't':x3,
    'ang': y3,
    'acc': z3,
})
#------------------------------N4-------------------------------------
x4=[]
y4=[]
z4=[]
with open('BaseDatos/N5.csv', 'r') as csvfile:
    plots4= csv.reader(csvfile, delimiter=',')
    for row4 in plots4:
        x4.append(int(row4[0]))
        y4.append(int(row4[1]))
        z4.append(int(row4[2]))
d4 = pd.DataFrame({
    't':x4,
    'ang': y4,
    'acc': z4,
})        
```
